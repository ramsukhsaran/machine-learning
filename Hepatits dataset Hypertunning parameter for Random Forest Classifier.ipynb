```python
import pandas as pd
import numpy as np                            
from matplotlib import pyplot as plt 


df=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/hepatitis/hepatitis.data")
#reading csv file from uci 
```

```python
col_name=['CLASS','AGE','SEX','STEROID','ANTIVIRAL','FATIGUE','MALAISE','ANOREXIA','LIVER BIG','LIVER FIRM','SPLEEN PALPABLE','SPIDERS','ASCITES','VARICES','BILIRUBIN','ALKPHOSPHATE','SGOT','ALBUMIN','PROTIME','HISTOLOGY']                               
```

```python
#setting columns name for data 
```

```python
df.columns =col_name        
  # or #df.set_axis(col_name, axis=1, inplace=True)                                 
```

```python
df.head()                                                                     # to view are datasets
```

```python
#building Feature set
print("No of featurs:",df.shape[1])
```

```python
print("shape of data",df.shape)
```

```python
print("no of unique class in dataste:",df['CLASS'].nunique())
```

```python
#Handling missing values in dataset 
```

```python
rr={'?':np.nan}      # in dataset we have  ?(not recoginze by machine) we replace with NAN value

df.replace(rr,inplace=True)

print("values ? convetrted into NAN")
df.head()
```

```python
col_mask=df.isnull().any(axis=0)
print(col_mask)                                  # to ckeck how many column have missing value(NAN)
```

```python
#to ckeck datatype of columns
df.dtypes

# some column are object type we need to convert it into numeric because machine learning model work with numeric values only
```

```python
df=df.astype(float)
print("Object are changed into numeric")
print(df.dtypes)


                                            # print(df.median()) it will print median value for each comlumn

# filing missing values  with median

df.fillna(df.median(), inplace=True)         
df.head(20)           
```

```python
#seprating target and input variable from dataframe

y=df['CLASS']   #target variable 

x=df.drop('CLASS',axis=1)  #input varuable or training data 

#input data without target variable thats why i drop first columns of dataset and assign to x

print("shape of y:",y.shape)
print("shape of x:",x.shape)
```

```python
# spliting training and testing data with help of scikit learn
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
print("Original data x:",x.shape)
print("After spliting x_traing shape:",np.shape(x_train))
print("After spliting x_test size:",np.shape(x_test))
```

```python
#Evaluate the accuracy of random forest  with number of estimator n=3

from sklearn.ensemble import RandomForestClassifier
tree=RandomForestClassifier(criterion='gini',max_depth=2,n_estimators=3,random_state=0)

                                                     
tree.fit(x_train,y_train)                                                                                 # put data orignal scale
y_pred=tree.predict(x_test)

from sklearn.metrics import accuracy_score

print('misclassified samples: %d'%(y_test!=y_pred).sum())
print("accuracy:%f"%accuracy_score(y_test,y_pred))

```

```python
# to find best parameter we need to do hyperparameter tunning 
```

```python
from sklearn.model_selection import GridSearchCV

#setup the parametrs and distribution to sample 
param_dist={'n_estimators':[5,10,15,20,],'max_features':['auto','sqrt'],'max_depth':[3,None],'min_samples_leaf':np.arange(1,10),"criterion": ["gini", "entropy"],'bootstrap':[True,False]}
rfc=RandomForestClassifier()
cv=GridSearchCV(rfc,param_dist,cv=5)
cv.fit(x_train,y_train)
y_p=cv.predict(x_test)

print("tuned parametrs:{}".format(cv.best_params_))
print("best Score is: {}".format(cv.best_score_))
```
